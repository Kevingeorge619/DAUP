```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Alphabet Stock.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# --- Define the date range ---
start_date = "2020-01-01"
end_date   = "2020-12-31"

# Filter data between two dates
mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
data = df.loc[mask]
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['Date'], data['Close'], label='Closing Price', color='blue')
plt.title("Alphabet Inc. - Historical Stock Prices")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(data['Date'], data['Volume'], color='orange')
plt.title("Alphabet Inc. - Trading Volume")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(
    [data['Open'], data['Close'], data['High'], data['Low']], 
    bins=40,  # more bins
    stacked=True, 
    label=['Open', 'Close', 'High', 'Low']
)
plt.title("Stacked Histogram of Stock Prices")
plt.xlabel("Price (USD)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['Volume'], data['Close'], alpha=0.6, c='green')
plt.title("Trading Volume vs Stock Price")
plt.xlabel("Volume")
plt.ylabel("Closing Price (USD)")
plt.show()
```

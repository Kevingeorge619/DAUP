```python
# Matplotlib is a low level graph plotting library in python that serves as a visualization utility.
# Matplotlib is open source and can  be use it freely
# install the library using pip install matplotlib
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Draw a line in a diagram from position (1, 3) to position (8, 10):

x = [1,8]
y = [3,10]
plt.plot(x,y)
plt.show()
```

```python
# plotting without line

x = [1,2,3,4,5,6]
y = [1,5,3,5,7,8]
plt.plot(x,y,'*') # marks the point with marker *
plt.show()
```

```python
y = [1,5,3,5,7,8] # Y value, if we dont specify X values by default it will start from 0 onwards
plt.plot(y)
plt.show()
```

```python
# matplotlib markers
import numpy as np
y= np.array([3,8,1,10])
plt.plot(y, marker = '*') # mark each pount with a star
plt.show()
```

```python
import numpy as np
x = np.arange(0,10,0.1) # x will be having value ranging from 0-10 with a step value of 0.1
y = np.sin(x)  # generating list y, containing the sin value of each x value in the list

# Create label and title for a plot
plt.plot(x,y)
plt.xlabel('angle') # setting label
plt.ylabel('sin')
plt.title('sin wave') # setting title for the plot
plt.show()
```

```python
# plot multiple lines
y = [1,2,3,4,5]
x = [3,3,3,3,3]
plt.plot(x,y, label = "line1", linestyle = "-")
plt.plot(y,x, label = "line2", linestyle = "--")
plt.plot(x, np.sin(x), label = "curve1")
plt.plot(x, np.cos(x), label = "curve2")
plt.legend()
plt.show()
```

```python
#display only grid lines
x=np.array([2,4,6,8,10])
y=np.array([6,7,2,4,7])
plt.plot(x,y)
plt.grid(axis='x')
plt.grid(axis='y')
plt.show()
```

```python
# display multiple plot - with subplot() function you can draw mulitple plots in one figure

#plot 1:
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
plt.subplot(2,2,1)
plt.plot(x,y)

#plot 2:
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])
plt.subplot(2,2,2)
plt.plot(x,y)

#plot 3:
x = np.array([0,1,2,3])
y =np.array([60,70,80,90])
plt.subplot(2,2,3)
plt.plot(x,y)

plt.show()
```

```python
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])
fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,0].set_title('Axis [0,0]')
axs[0,1].plot(x,y,'tab:orange')
axs[0,1].set_title('Axis[0,1]')
axs[1,0].plot(x,y,'tab:green')
axs[1,0].set_title('Axis [1,0]')
axs[1,1].plot(x,y, 'tab:red')
axs[1,1].set_title('Axis[1,1]')
plt.show()
```

```python
# Bar Plot

x = ['Physics','Chemistry', 'Maths']
y = [98, 85, 100]
plt.bar(x,y)
plt.show()
```
